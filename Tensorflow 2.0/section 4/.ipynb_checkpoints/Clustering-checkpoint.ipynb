```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=2000, centers=3, n_features=2,
                   random_state=0)
```

```python
X_test, y_test = make_blobs(n_samples=500, centers=3, n_features=2,
                   random_state=0)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, s=100, edgecolor="k", linewidth=2)
plt.show()
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, input_shape=(2,), activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
# Compile and Fit
opt = tf.keras.optimizers.Adam(0.01)
model.compile(optimizer=opt, loss='mse')
r = model.fit(X, y, validation_data=(X_test, y_test), epochs=100)
```

```python
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
XX, yy = make_blobs(n_samples=500, centers=3, n_features=2,
                   random_state=0)
```

```python
Yhat = model.predict(XX).flatten()
print("Final accuracy:", model.evaluate(XX, yy))
```
