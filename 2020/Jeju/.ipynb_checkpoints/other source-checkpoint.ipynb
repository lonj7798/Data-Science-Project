## ADV data

```python
ad_data = pd.read_csv("data/당월대비_익월_매체별_업종별_광고경기전망지수_20200712161649.csv", encoding="CP949")
ad_data.head()
```

```python
date = []

for i in ["2019", "2020"]:
    for j in range(1,13):
        if j < 10:
            temp = i + ". 0" + str(j)
            date.append(temp)
            if (i=="2020") & (j==4):
                break
        else:
            temp = i + ". " + str(j)
            date.append(temp)    
```

```python
data = pd.DataFrame(ad_data.loc[1, date])
data.index = pd.to_datetime(data.index)
data.rename(columns={1:"adv_data"}, inplace=True)
```

## GDP data

```python
gdp_data = pd.read_csv("data/경제활동별_GDP_및_GNI_원계열__실질__분기_및_연간__20200712154907.csv", encoding="CP949")
gdp_data
```

```python
# GDP, GDI, GNI 사용

temp = pd.DataFrame(gdp_data.iloc[[51, 53, 55], :])
temp= temp.T.iloc[1:, :].rename(columns={51:"GDP", 53:"GDI", 55:"GNI"})
temp
```

```python
data["GDP"] = [temp.GDP.iloc[0]]*4 + [temp.GDP.iloc[1]]*4 + [temp.GDP.iloc[2]]*4 + [temp.GDP.iloc[3]]*4
data["GDI"] = [temp.GDI.iloc[0]]*4 + [temp.GDI.iloc[1]]*4 + [temp.GDI.iloc[2]]*4 + [temp.GDI.iloc[3]]*4
data["GNI"] = [temp.GNI.iloc[0]]*4 + [temp.GNI.iloc[1]]*4 + [temp.GNI.iloc[2]]*4 + [temp.GNI.iloc[3]]*4
data
```
