### Seller vs Customer location

```python
# Seller Location
seller_loc = data.groupby("seller_location_1", as_index=False)[['amount']].sum().sort_values(by="amount", ascending=False)
labels_seller = seller_loc.seller_location_1
sizes_seller = seller_loc.amount

# Customer Location
customer_loc = data.groupby("customer_location_1", as_index=False)[['amount']].sum().sort_values(by="amount", ascending=False)
labels_customer = customer_loc.customer_location_1
sizes_customer = customer_loc.amount

# Graph
fig1, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 15))

ax[0].pie(sizes_seller, labels=labels_seller, autopct='%1.1f%%', shadow=True, startangle=90)
ax[0].set_title("Seller's location", size=16)

ax[1].pie(sizes_customer, labels=labels_customer, autopct='%1.1f%%', shadow=True, startangle=90)
ax[1].set_title("Customer's location", size=16)
```

```python
import statsmodels.api as sm
from pylab import rcParams

def plotseasonal(res, axes ):
    res.observed.plot(ax=axes[0], legend=False)
    axes[0].set_ylabel('Observed')
    res.trend.plot(ax=axes[1], legend=False)
    axes[1].set_ylabel('Trend')
    res.seasonal.plot(ax=axes[2], legend=False)
    axes[2].set_ylabel('Seasonal')
    res.resid.plot(ax=axes[3], legend=False)
    axes[3].set_ylabel('Residual')
```

```python
temp = data.groupby(["kind","seller_location_1", "date"], as_index=False)[['amount']].sum()
temp.date = pd.to_datetime(temp.date, format='%Y%m')
```

```python
def graph_seasonal(field):
    # graph 1
    data_1 = temp[temp['kind']==field]
    sns.catplot(x='date', y='amount', hue='seller_location_1', data=data_1, kind='point', aspect=2, palette='Set1')
    
    # graph 2
    data_2 = data_1.groupby("seller_location_1", as_index=False)[['amount']].sum().sort_values(by="amount", ascending=False)
    labels = data_2.seller_location_1
    sizes = data_2.amount

    fig1, ax = plt.subplots(figsize=(7, 7))
    ax.pie(sizes, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    
    # graph3
    data_3 = data_1.groupby("date")[['amount']].sum()
    train_amount= sm.tsa.seasonal_decompose(data_3.amount,period=4) # 4 mount

    fig, axes = plt.subplots(ncols=1, nrows=4, sharex=True, figsize=(15,7))
    plotseasonal(train_amount, axes[:])
    plt.show()
```
