### 공정 순서

PRT: 원형제작 \
MOL: 성형 \
BLK: 자르기

PRT_1 -> MOL_1 -> BLK_1 \
PRT_2 -> MOL_2 -> BLK_2 \
PRT_3 -> MOL_2 -> BLK_3 \
PRT_4 -> MOL_4 -> BLK_4

### PRT(원형제작) 특이사항 
- 하루 최대 500개 투입 가능 (PRT_1~4 총합)
- 시간당 6.66개 투입 가능
- 투입 일 후 원형 생산 완료
- 1개 라인으로 구성

### 파일제출
- PRT_1 ~ 4에 투입 개수 작성

-------------------------------

### MOL(성형) 특이사항
- 2개의 라인으로 구성 (A,B)
- 투입시 48 공정 시간 후 생산됨
- 양품률 97.5%
- 점검(CHECK), 공정(PROCESS), 변경(CHANGE), 멈춤(STOP)
- 점검 공정 변경 멈춤은 연달아 할 수 없읍
- 공정 중 시간당 최대 6.667개 투입 가능
- 날자별 최대 투입 개수가 다름 (max_count.csv 참고)


### 제약 사항
- 공정 전 점검 필수 28시간 소요
- 점검 후 최소 98시간, 최대 140시간 공정 가동
- 공정 중 변경가능, 변경시 일정 시간 소요 (change_time.csv 참고)
- 점검 중 아이템 변경 가능, 28시간 소요
- 점검 시간보대 갈게 공정 중단시, 쵀대 8일 까지 정지(STOP)가능, 정지 후 점검 필수

------------------------------

### MOL(성형) 특이사항
- PROCESS: 공정
- STOP: 멈춤
- CHECK: 점검, 이 후 투입되는 양 만큼 공정 시간 동안 MOL생산 \
    - 숫자에 따라 MOL_1 ~ 4 생산
- CHANGe: 변경, 이후 공정에 투입되는 양 만큼 MOL 생산
    - 숫자에 따라 MOL_1 ~ 4 생산
    - ex) CHANGE_12: 투입 아이템을 PRT_1에서 PRT_2로 변경
    
    
### 파일 제출
- Event_A, Event_B에 MOL_A, MOL_B 투입 개수 기입 (A, B: 각각의 라인)
- Event 종류: CHECK_1, CHECK_2, CHECK_3, CHECK_4,
             CHECK_11, CHECK_12, CHECK_13, CHECK_14,
             CHECK_21, CHECK_22, CHECK_23, CHECK_24,
             CHECK_31, CHECK_32, CHECK_33, CHECK_34,
             CHECK_41, CHECK_42, CHECK_43, CHECK_44,
             PROCESS, STOP
- 투입 개수(MOL_A, MOL_B)는 PROCESS 일때만 0 이상 기입 가능

### BLK(자르기) 특이사항
- 수요 발생 시 성형이 완료된 아이템을 모두 자르고 납품
- 양품률: cut_yield.csv
- MOL_1, MOL_2 -> 506 BLK_1, BLK_2
- Mol_3, MOL_4 -> 400 BLK_3, BLK_4
- 납품 시간: 오후 6시
- 자르기는 자동으로 계산됨 (제출 X)
