```python
import pandas as pd
import numpy as np

# 지도에 표시
import folium
from folium.features import DivIcon
```

```python
df = pd.read_csv("data/a1.txt", sep='\s+', header=None)
df = df.loc[:, 2:6]
df.drop([4,5], axis=1, inplace=True)

df.columns=['위도', '경도', 'time']
df.time = pd.to_datetime(df.time)
df.index = df.time
```

```python
temp = df.resample('1T').mean()
```

```python
map_a = folium.Map(location=[temp.위도.mean(), temp.경도.mean()], zoom_start=100)
```

```python
for i in range(0, len(temp)):
    
    name = '<div style="font-size: 18pt; color : black">%s</div>' %str(i)
    p = [temp.iloc[i].위도, temp.iloc[i].경도]
    folium.Marker(p, icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(7,20),
            html=name
            )).add_to(map_a)
    map_a.add_child(folium.CircleMarker(p, radius=10))

map_a
```

```python
for i in range(0, len(temp)):
    
    folium.CircleMarker([temp.iloc[i].위도, temp.iloc[i].경도], radius=10, 
                            color='green',).add_to(map_a)
    
map_a
```

```python
m = folium.Map(location=[temp.위도.mean(), temp.경도.mean()], tiles='cartodbpositron', zoom_start=30)
m

# titles: kind of map
```

```python
folium.RegularPolygonMarker(location=[temp.위도.mean(), temp.경도.mean()], 
                            fill_color='blue', number_of_sides=3, 
                            radius=10, rotation=0).add_to(m)
m
```

```python
my_PolyLine=folium.PolyLine(locations=temp.values,weight=1)
m.add_children(my_PolyLine)
```
