```python
import pandas as pd
import numpy as np

import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import datetime as dt
```

```python
train = pd.read_csv("../../Data/Price prediction/train.csv")
test = pd.read_csv("../../Data/Price prediction/test.csv")
```

```python
print("shape of the train:", train.shape)
print("shape of the test:", test.shape)
# 8개의 feature
```

```python
# datetime 타입으로 변경
# is_calceled feature 추가

combine = [train, test]

for data in combine:
    data['is_canceled'] = data['amount'].map(lambda x: 1 if x < 0 else 0)
    data['date'] = pd.to_datetime(data['date'])
    
data = pd.concat([train, test]).sort_values(by='store_id')
```

### 연도별 가게 수 비교

```python
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month

temp = data.groupby(['year', 'store_id']).mean()
temp2 = data.groupby(['year', 'month', 'store_id']).mean()
```

```python
# monthly_df = data.resample('M').size()
```

```python
# 2016/08 ~ 2018/07

year_month = []
number_of_stores = []

for i in range(8, 13):
    if i < 10:
        name = "2016-0" + str(i)
        year_month.append(name)
        
        num = len(temp2.xs(2016).xs(i))
        number_of_stores.append(num)
    else:
        name = "2016-" + str(i)
        year_month.append(name)
        
        num = len(temp2.xs(2016).xs(i))
        number_of_stores.append(num)
##        
for i in range(1, 13):
    if i < 10:
        name = "2017-0" + str(i)
        year_month.append(name)
        
        num = len(temp2.xs(2017).xs(i))
        number_of_stores.append(num)
    else:
        name = "2017-" + str(i)
        year_month.append(name)
        
        num = len(temp2.xs(2017).xs(i))
        number_of_stores.append(num)
##        
for i in range(1, 8):
    name = "2018-0" + str(i)
    year_month.append(name)
        
    num = len(temp2.xs(2018).xs(i))
    number_of_stores.append(num)
        
year_month_df = pd.DataFrame({'year_month': year_month, 'number_of_stores': number_of_stores})
```

```python
# 월별 상점 수

plt.figure(figsize=(16,5))
plt.title("Number of Stores: Month")
sns.pointplot(x="year_month", y='number_of_stores', data=year_month_df, ci=None)
```

```python
# 거래 횟수 대비 slaes

store = data.groupby(['year', 'store_id']).sum()[['amount' , 'is_canceled']]
store2 = data.groupby(['year', 'store_id']).size()

store['trades'] = store2
store['sales_per_trade'] = store['amount'] / store['trades']
```

```python
store_2016 = store.xs(2016).sort_values('sales_per_trade', ascending=False)
store_2017 = store.xs(2017).sort_values('sales_per_trade', ascending=False)
store_2018 = store.xs(2018).sort_values('sales_per_trade', ascending=False)
```

```python
# 전체 sales / trades
store = data.groupby('store_id')[['amount', 'is_canceled']].sum()
store2 = data.groupby('store_id').size()

store['number_of_trades'] = store2
store['sales_per_trade'] = store['amount'] / store['number_of_trades']
store.sort_values(by='sales_per_trade', inplace=True, ascending=False)
```

```python
store.tail(10)
# 비정상적인 6개를 제외한 나머지 비교
```

```python
# 매출액 대비 거래 횟수

print("상위 100개 상점의 총 매출액:", store['amount'].head(100).sum())
print("하위 143개 상점의 총 매출액:", store['amount'].tail(143).sum())

print("상위 100개 상점의 매출액 차지: ", store['amount'].head(100).sum()/store['amount'].sum() )
# 약 4.7% 차지.
```

```python
# 매출액 비교

store.sort_values(by='amount', ascending=False, inplace=True)
print("상위 50개 상점의 총 매출액:", store['amount'].head(50).sum())
print("하위 1200개 상점의 총 매출액:", store['amount'].tail(1200).sum())
print("상위 50개 상점의 매출액 차지:", store['amount'].head(50).sum() / store['amount'].sum())

# 상위 50개의 상점이 총 매출액의 25퍼센트를 차지.
```
